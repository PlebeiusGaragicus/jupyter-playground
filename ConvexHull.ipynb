The first step in designing an algorithm is to understand the problem you want to solve. Let’s start with a sample problem from the field of computational geometry. Given a set of points, P, in a two-dimensional plane, such as shown in Figure 1-1, picture a rubberband that has been stretched around the points and released. The resulting shape is known as the **convex hull** (i.e., the smallest convex shape that fully encloses all points in P). ***Your task is to write an algorithm to compute the convex hull from a set of two-dimensional points.***

```
import matplotlib.pyplot as plt
import numpy as np
import random as r
def plotPoints( ps, char, _color="black"):
    for p in ps:
        x, y = p
        plt.plot( x, y, char, color=_color )
def slope(p1, p2):
    try:
        return ( p2[1] - p1[1] ) / ( p2[0] - p1[0] )
    except:
        return 0
def findNext_q1( lastP ):
    searchBox = []
    for p in pts:
        if p[1] > lastP[1]:
            searchBox.append( p )
    nextP = searchBox[0]
    for p in searchBox:
        if slope(p, lastP) > slope(nextP, lastP):
            nextP = p
    return nextP
def findNext_q2( lastP ):
    searchBox = []
    for p in pts:
        if p[0] > lastP[0]:
            searchBox.append( p )
    nextP = searchBox[0]
    for p in searchBox:
        if abs(slope(p, lastP)) < abs(slope(nextP, lastP)):
            nextP = p
    return nextP
def findNext_q3( lastP ):
    searchBox = []
    for p in pts:
        if p[1] < lastP[1]:
            searchBox.append( p )
    nextP = searchBox[0]
    for p in searchBox:
        if slope(p, lastP) > slope(nextP, lastP):
            nextP = p
    return nextP
def findNext_q4( lastP ):
    searchBox = []
    for p in pts:
        if p[0] < lastP[0]:
            searchBox.append( p )
    nextP = searchBox[0]
    for p in searchBox:
        if abs(slope(p, lastP)) < abs(slope(nextP, lastP)):
            nextP = p
    return nextP
numPts = 100
pts = [ (0,0) ]
for i in range( numPts - 1 ):
    rx = (r.random() * 20) - 10
    ry = (r.random() * 20) - 10
    newp = (round(rx, ndigits=2), round(ry, ndigits=2))
    pts.append( newp )
plotPoints( pts, '.', "black")
minX = pts[0]
minY = pts[0]
maxX = pts[0]
maxY = pts[0]
for p in pts[1:]:
    if p[0] < minX[0]:
        minX = p
    if p[1] < minY[1]:
        minY = p
    if p[0] > maxX[0]:
        maxX = p
    if p[1] > maxY[1]:
        maxY = p
edgeP = [ minX ]
while True:
    if edgeP[-1] == maxY:
        break
    found = findNext_q1( edgeP[-1] )
    edgeP.append( found )
while True:
    if edgeP[-1] == maxX:
        break
    found = findNext_q2( edgeP[-1] )
    edgeP.append( found )
while True:
    if edgeP[-1] == minY:
        break
    found = findNext_q3( edgeP[-1] )
    edgeP.append( found )
while True:
    if edgeP[-1] == minX:
        break
    found = findNext_q4( edgeP[-1] )
    edgeP.append( found )
xs = []
ys = []
for p in edgeP:
    xs.append( p[0] )
    ys.append( p[1] )
xs.append( minX[0] )
ys.append( minX[1] )
plt.plot(xs, ys,'b-')
plt.show()
```
